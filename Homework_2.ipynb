# Task_2

```python
t = np.linspace(0, 10, 51)
t
```

```python
f = np.cos(t)
f
```

```python
plt.plot(t, f, color = 'green')
# Title
plt.title('Graph f(t)', fontsize = 16)
# Axis X.title
plt.xlabel('t')
# Axis Y.title
plt.ylabel('Graph f(t)')
# Set limit in range axis X & Y
plt.xlim([0.5, 9.5])
plt.ylim([-2.5, 2.5])

plt.show()
```
